```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# New Antecedent/Consequent objects hold universe variables and membership
# functions
distance = ctrl.Antecedent(np.arange(0, 351, 1), 'distance')
motor_1 = ctrl.Consequent(np.arange(0, 256, 1), 'motor_1')
motor_2 = ctrl.Consequent(np.arange(0, 256, 1), 'motor_2')

motor_1.defuzzify_method = 'mom'
motor_2.defuzzify_method = 'mom'

# Custom membership functions can be built interactively with a familiar,
# Pythonic API
distance['very close'] = fuzz.trapmf(distance.universe, [0, 0, 40, 70])
distance['close'] = fuzz.trapmf(distance.universe, [40, 70, 120, 150])
distance['normal'] = fuzz.trapmf(distance.universe, [120, 150, 200, 230])
distance['far'] = fuzz.trapmf(distance.universe, [200, 230, 280, 310])
distance['very far'] = fuzz.trapmf(distance.universe, [280, 310, 350, 351])

# Output membership functions 
motor_1['medium'] = fuzz.gbellmf(motor_1.universe, 60, 3, 0)
motor_1['high'] = fuzz.gbellmf(motor_1.universe, 60, 3, 127)
motor_1['very high'] = fuzz.gbellmf(motor_1.universe, 60, 3, 256)

motor_2['medium'] = fuzz.gbellmf(motor_2.universe, 60, 3, 0)
motor_2['high'] = fuzz.gbellmf(motor_2.universe, 60, 3, 127)
motor_2['very high'] = fuzz.gbellmf(motor_2.universe, 60, 3, 256)
```

```python
# You can see how these look with .view()
distance.view()
```

```python
motor_1.view()
```

```python
# print(help(motor_1['high'] & motor_2['medium']))
rule1 = ctrl.Rule(distance['very close'], (motor_1['very high'], motor_2['medium']))   
rule2 = ctrl.Rule(distance['close'], (motor_1['high'], motor_2['medium']))
rule3 = ctrl.Rule(distance['normal'], (motor_1['medium'], motor_2['medium']))
rule4 = ctrl.Rule(distance['far'], (motor_1['medium'],motor_2['high']))
rule5 = ctrl.Rule(distance['very far'], (motor_1['medium'], motor_2['very high']))
print(type(motor_1['very high'] & motor_2['medium']))
```

```python
motion_control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
```

```python
motion = ctrl.ControlSystemSimulation(motion_control)
```

```python
# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
motion.input['distance'] = 135
# Crunch the numbers
motion.compute()
```

```python
motor_2.view(sim=motion)
print(motion.output['motor_2'])
```

```python
motor_1.view(sim=motion)
print(motion.output['motor_1'])
```
